# Simple Linear regression step-by-step

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([5,7,12,16,20])
y=np.array([40,120,180,210,240])
```

```python
def regression(myM, myX, myB):
    return myM*myX+myB  #return myhhats
```

```python
#free initialization of the coefficients m and b
m=0.9
b=0.5
```

```python
#calcolo y0
yPred = regression(m, x, b)
print(yPred)
```

```python
#define my cost function MSE
def costfunction(yhatslice, y):
    MSE = np.square(np.subtract(y, yhatslice)).mean()
    return MSE
```

```python
#derivative dC/dm
#1/n*(y-yhat)**2=2/n*m
def gradient_m(yhats,y_true,x):
    dcdm=2*1/len(y_true)*np.sum((yhats-y_true)*x)
    return dcdm
```

```python
#derivative dC/db
def gradient_b(yhats,y_true):
    dcdb = 2*1/len(y_true)*np.sum((yhats-y_true))
    return dcdb
```

```python
def gradient(yPred,y,x):
    dm=gradient_m(yPred,y,x)
    db = gradient_b(yPred,y)
    return [dm,db]
```

```python
#SGD
lr=0.01 #learning rate
scaled_gradient = lr * np.array(gradient(yPred,y,x))
scaled_gradient
```

```python
#newm=m-lr*dC/dm
newm = m - np.array(gradient(yPred,y,x))[0]
#newm=m-lr*dC/db
newb = b - np.array(gradient(yPred,y,x))[1]
print(newm, newb)
```

```python
epochs=15
m=0.9
b=0.5
lr=0.01
x=np.array([0, 1, 2, 3, 4, 5, 6, 7.])
y=np.array([1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37])
for r in range(epochs):
    yhats = regression(m,x, b)
    C = costfunction(yhats, y)

    print('Step {}, cost {}, m grad {}, b grad {}, lr {}'.format(r, '%.3g' % C, '%.3g' % m, '%.3g' % b, '%.3g' % lr))
    gradvect=gradient(yhats,y,x)
    theta=np.array([m,b])
    new_theta = theta - lr * np.array(gradvect)
    m=new_theta[0]
    b=new_theta[1]
```

```python
#we can add plots #thanks to john krhon
def labeled_regression_plot(my_x, my_y, my_m, my_b, my_C):
    title = 'Cost = {}'.format('%.3g' % my_C)
    xlabel = 'm = {}, m grad = {}'.format('%.3g' % my_m, '%.3g' % my_m)
    ylabel = 'b = {}, b grad = {}'.format('%.3g' % my_b, '%.3g' % my_b)
    fig, ax = plt.subplots()

    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)

    ax.scatter(my_x, my_y)

    x_min, x_max = ax.get_xlim()
    y_min = regression(my_m,x_min, my_b)
    y_max = regression(my_m,x_max, my_b)

    ax.set_xlim([x_min, x_max])
    _ = ax.plot([x_min, x_max], [y_min, y_max], c='C01')
```

```python
epochs=15
m=0.9
b=0.5
lr=0.01
x=np.array([0, 1, 2, 3, 4, 5, 6, 7.])
y=np.array([1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37])
for r in range(epochs):
    yhats = regression(m,x, b) 
    C = costfunction(yhats, y)
    labeled_regression_plot(x, y, m, b, C)
    gradvect=gradient(yhats,y,x)
    theta=np.array([m,b])
    new_theta = theta - lr * np.array(gradvect)
    m=new_theta[0]
    b=new_theta[1]
```
